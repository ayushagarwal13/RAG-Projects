## Chunking

Interview Question

If the interviewer asks:

Why do we split a document into chunks?

A strong answer is:

"We split a document into chunks because LLMs have context window limits, and sending an entire document is slow, expensive, and often unnecessary. Chunking allows us to retrieve only the most relevant sections of the document, improving both efficiency and response quality."

If the interviewer asks:

Does one chunk represent one page?

You should answer:

"Not necessarily. In most RAG systems, chunks are created based on a fixed character or token size with some overlap to preserve context. A chunk may contain part of a page, an entire page, or even span across multiple pages depending on the configured chunk size."

## Embeddings

If the interviewer asks:

What is an embedding?

A strong answer is:

"An embedding is a high-dimensional numerical vector that represents the semantic meaning of a piece of text. Texts with similar meanings have embeddings that are close to each other in vector space, enabling semantic search instead of exact keyword matching."

cosine similarity compares vectors not texts. Cosine similarity simply measure how close those vectors are.

How does semantic search work?

A strong answer is:

"Both the document chunks and the user's query are converted into embeddings using the same embedding model. The vector database then computes cosine similarity between the query embedding and the stored chunk embeddings. The chunks with the highest similarity scores are retrieved and passed to the LLM."

Embedding models

What happens in Phase 2?

Suppose the user asks:

"How many casual leaves are given?"

Your system will do this:

User Question
       │

       │

Notice something:

Gemini is still NOT involved.

We're only testing whether our vector database can find the correct information.

```python
#Step 1: Create a Retriever
#Your vector_db already exists.
#Now convert it into a retriever.
retriever = vector_db.as_retriever(
    search_kwargs={"k": 3}
)
# think chromadb as a library and retirieval is an librarian
```

```python
#Step 2: Ask a Question
query = "How many casual leaves are given?"
```

```python
#Step 3: Retrieve Similar Chunks
results = retriever.invoke(query)
```

We are passing plain text, not embeddings.

Why?

Because the retriever automatically:

Converts the question into an embedding.
Searches ChromaDB.
Finds the nearest vectors.
Returns the original text.

```python
#Step 4: See what was retrieved
print(results)
# the output will be 4 document because langchain uses a default value called k=4 meaning return the top 4 most similar chunks
```
